## Model building (by Classification)

```python
df2=dfWtr.values
```

```python
df3=pd.DataFrame(df2)
```

```python
df4=df3.fillna(0)
print(df4)
```

```python
#divide data into feature matrix and target matrix
X = df2[:,:4] # all the rows and first four columns 0-3
Y = df2[:,9]   # all the rows and only ninth column (last column )
print(X.shape, Y.shape)
```

```python
print(X)
```

```python
print(Y)
```

```python
X[np.isnan(X)] = 0
```

```python
Y[np.isnan(Y)] = 0
```

```python
from sklearn.linear_model import LogisticRegression
model= LogisticRegression()
```

```python
model.fit(X,Y)
```

## Predicting Potability of Water
